One Hot Encoding

```python
import pandas as pd
```

```python
df_dum=pd.get_dummies(df[['sex','smoker','region']],drop_first=True,dtype='int')
```

```python
df_dum
```

```python
df_num=df.drop(columns=['sex','smoker','region'],axis=1)
df_num
```

```python
df_total=pd.concat([df_dum,df_num],axis=1)
```

```python
df_total
```

```python
df_total['age_smoker']=df_total['age']*df_total['smoker_yes']
df_total['bmi_smoker']=df_total['bmi']*df_total['smoker_yes']
```

```python
df_total.head()
```

```python
relation=['age','bmi','children','age_smoker','charges','bmi_smoker','smoker_yes']
corr=df_total[relation].corr()
corr
plt.figure(figsize=(10,5))
sns.heatmap(corr,annot=True)
plt.tight_layout()
```

```python
x=df_total[['sex_male','smoker_yes','region_northwest','region_southeast','region_southwest','age','bmi','children','age_smoker','bmi_smoker']].values
x
```

```python
y=df['charges'].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_train
```

```python
x_test=scaler.transform(x_test)
x_test
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
from sklearn.model_selection import cross_val_score
validation_score=cross_val_score(model,x_train,y_train,scoring='neg_mean_squared_error',cv=3)
validation_score
```

```python
np.mean(validation_score)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

mse=mean_squared_error(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
rmse=np.sqrt(mse)
print(f"mse:{mse}")
print(f"mae:{mae}")
print(f"rmse:{rmse}")
```

```python
from sklearn.metrics import r2_score
r2=r2_score(y_test,y_pred)
r2
```

```python
plt.scatter(y_test,y_pred,color='green')
```

```python
residual=y_test-y_pred
residual
```

```python
sns.kdeplot(residual,color='orange',fill=True)
```

```python
import pickle

# This code assumes your final trained model variable is named 'model'
# and your StandardScaler is named 'scaler' from the earlier steps.

pickle.dump(model, open('model.pkl', 'wb'))
pickle.dump(scaler, open('scaler.pkl', 'wb'))

print("model.pkl and scaler.pkl have been saved successfully!")
```
